```python
from datetime import datetime, date
from matplotlib import pyplot as plt
import numpy as np
from bcb import sgs
```

```python
#Passo 2 - Coletar dados do usuário

capital = float(input("Digite o capital investido: "))
frequencia = input("Digite a frequência do período (Y, ME, D): ")
inicio = input("Digite a data inicial maior que 1995/01/01 no formatao YYYY/MM/DD: ")
final = input("Digite a data final no seguinte formatao YYYY/MM/DD: ")
```

```python
#Passo 3 - Tratar dados coletados
data_inicio = datetime.strptime(inicio, "%Y/%m/%d").date()
data_final = datetime.strptime(final, "%Y/%m/%d").date()

data_inicio
data_final
```

```python
#Passo 4 - Pegar dados da SELIC do banco central 
```

```python
#para pegar o numero da taxa selic acessar o site https://www3.bcb.gov.br/sgspub/
taxas_selic = sgs.get({"selic": 11}, start = data_inicio, end = data_final)

taxas_selic = taxas_selic / 100

taxas_selic
```

```python
#Passo 5 - Calcular retorno do período
```

```python
capital_acumulado = capital * (1 + taxas_selic["selic"]).cumprod() - 1

capital_acumulado
```

```python
capital_com_frequencia = capital_acumulado.resample(frequencia).last()

capital_com_frequencia
```

```python
#Parte 2

#Passo 1 - Filtrar dados da selic no periodo em questão
```

```python
data_inicio_2 = date(2000, 1 ,1)
data_final_2 = date(2022, 3, 1)
```

```python
selic_questao_2 = sgs.get({"selic": 11}, start = data_inicio_2, end = data_final_2) / 100

selic_questao_2
```

```python
#Passo 2 - Calcular rentabilidade das janelas de 500 dias
```

```python
janelas_500_dias = ((1 + selic_questao_2).rolling(window = 500).apply(np.prod) - 1)

janelas_500_dias
```

```python
janelas_500_dias = janelas_500_dias.reset_index()

janelas_500_dias
```

```python
#Passo 3 - Crias range de datas na tabela
```

```python
janelas_500_dias["data_inicial"] = janelas_500_dias["Date"].shift(500)

janelas_500_dias
```

```python
janelas_500_dias = janelas_500_dias.dropna()

janelas_500_dias
```

```python
janelas_500_dias.columns = ["data_final", "retorno_selic_500d", "data_inicial"]

janelas_500_dias
```

```python
#Passo 4 - Pegar o maior retorno da tabela 
```

```python
maior_retorno = janelas_500_dias["retorno_selic_500d"].max()

maior_retorno
```

```python
gabarito = janelas_500_dias[janelas_500_dias["retorno_selic_500d"] == maior_retorno]

gabarito
```
